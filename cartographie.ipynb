```python
import geopandas as gpd
sf = gpd.read_file('departements-version-simplifiee.geojson')
sf['code'] = sf['code'].astype(str)
sf.head()
sf.dtypes
```

```python
import sys 
df = df.dropna(subset=['Code postal'])
df = df.dropna(subset=['Surface terrain'])



def tronquer_valeur(valeur):
    str_valeur = str(valeur)
    if len(str_valeur) == 7:
        return int(str_valeur[:2])
    elif len(str_valeur) == 6:
        return int(str_valeur[:1])
    else: 
        return str_valeur

def ajout_0(valeur):
    str_valeur = str(valeur)
    if len(str_valeur) == 1:
        return ('0' + str(str_valeur))
    else: 
        return str_valeur
    
df['Code postal'] = df['Code postal'].astype(str)
df['Code postal'] = df['Code postal'].apply(tronquer_valeur)
df['Code postal'] = df['Code postal'].apply(ajout_0)
df['Code postal'] = df['Code postal'].astype(str)


    
#dd = df.groupby('Code postal')['Valeur fonciere'].mean()
#print(df)

df
#df.dtypes
```

```python
sf = pd.merge(sf, df, left_on='code', right_on='Code postal')
sf
```

```python
sf['valeurm2']=  sf['Valeur fonciere']/sf['Surface reelle bati'] 
sf['valeurm2'] = sf['valeurm2'].astype(float)
sf['value'] = sf.groupby('code')['valeurm2'].mean()


res = sf.groupby('code')['valeurm2'].mean()
res.values    


#sf
```

```python
import numpy as np
from geoviews import dim
```

```python

#sf['value'] = sf['value'].astype(int)
#sf.dtypes
sf.drop_duplicates('code', inplace=True)
sf['value'] = res.values
```

```python
deps = gv.Polygons(sf, vdims=['nom','value'])

```

```python
from geoviews import dim
cartomaison = deps.opts(width=600, height=600,toolbar='above', color=dim('value'),
        colorbar=False, tools=['hover'], aspect='equal')
 
cartomaison

```
