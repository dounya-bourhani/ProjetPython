```python
import pandas as pd
chomage = pd.read_excel('sl_etc_2023T2.xls',sheet_name='Département')

total_rows = len(chomage)
chomage = chomage.iloc[:total_rows - 4]
chomage
```

```python
colonnes_selectionnees = chomage.filter(like='2018').columns.tolist() + \
                         chomage.filter(like='2019').columns.tolist() + \
                         chomage.filter(like='2020').columns.tolist() + \
                         chomage.filter(like='2021').columns.tolist()

nouveau_chomage = chomage[colonnes_selectionnees]
nouveau_chomage
```

```python
moyenne_par_ligne = nouveau_chomage.mean(axis=1)
moyenne_par_ligne
```

```python
code_postal = chomage['Code'].astype('str')
code_postal
```

```python
data = {
    'Code postal': code_postal,
    'Moyenne Taux Chomage': moyenne_par_ligne
}

moyenne_taux_chomage = pd.DataFrame(data)

print(moyenne_taux_chomage)
```

```python
import pickle

with open('dataframe.pkl', 'rb') as file:
    df = pickle.load(file)
```

```python
df.shape
```

```python
df = df.dropna(subset=['Code postal'])
# df = df.dropna(subset=['Surface terrain'])


def tronquer_valeur(valeur):
    str_valeur = str(valeur)
    if len(str_valeur) == 7:
        return int(str_valeur[:2])
    elif len(str_valeur) == 6:
        return int(str_valeur[:1])
    else: 
        return str_valeur

def ajout_0(valeur):
    str_valeur = str(valeur)
    if len(str_valeur) == 1:
        return ('0' + str(str_valeur))
    else: 
        return str_valeur
    
df['Code postal'] = df['Code postal'].astype(str)
df['Code postal'] = df['Code postal'].apply(tronquer_valeur)
df['Code postal'] = df['Code postal'].apply(ajout_0)
df['Code postal'] = df['Code postal'].astype(str)
```

```python
df_merged = df.merge(moyenne_taux_chomage, on='Code postal', how='left')
df_merged
```

```python
print("Pourcentage de ligne sans Taux de chomage :", df_merged['Moyenne Taux Chomage'].isna().sum() / df_merged.shape[0] * 100)
```

```python
df_merged['Valeur fonciere']
```

```python
df_merged['prix_par_m2'] = df_merged['Valeur fonciere'] / df_merged['Surface reelle bati']
```

```python
pd.to_numeric(df_merged['Valeur fonciere'], errors='coerce')
```

```python
grouped_values = df_merged.groupby('Code postal')['Valeur fonciere'].mean()

df_merged['moyenne_prix_par_m2_par_code_postal'] = df_merged['Code postal'].map(grouped_values)  # Ajoute la moyenne par 'Code postal'
df_merged
```

```python
# Supprimer les lignes où 'Nature mutation' n'est pas 'vente'
df_merged = df_merged.loc[df_merged['Nature mutation'] == 'Vente']

# Vérifier que les lignes ont été supprimées
print(df_merged['Nature mutation'].value_counts())
```

```python
proportions = df_merged['Nature mutation'].value_counts(normalize=True)
print(proportions)
```
